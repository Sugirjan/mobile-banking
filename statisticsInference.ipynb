```
def t_test(x,y,alternative='both-sided'):
            _, double_p = stats.ttest_1samp(x,y)
            if alternative == 'both-sided':
                pval = double_p
            elif alternative == 'greater':
                if np.mean(x) > np.mean(y):
                    pval = double_p/2.
                else:
                    pval = 1.0 - double_p/2.
            elif alternative == 'less':
                if np.mean(x) < np.mean(y):
                    pval = double_p/2.
                else:
                    pval = 1.0 - double_p/2.
            return _, pval
cols = [ 'TimeSaving', 'Easyness', 'UI_Design', 'Network_Connectivity', 'Cost_effective', 'Availabilty','Security', 'Comfortable', 'Awareness', 'Usage', 'Satifaction']
for col in cols:
  print(col, t_test(df_scaled[df_scaled.Satifaction.notnull()][col],3.0, 'greater'))
```
